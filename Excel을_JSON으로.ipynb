```python
import pandas as pd
import json

# 1. 엑셀 파일 읽기
df = pd.read_excel('/content/drive/MyDrive/8팀/음식-술-약물 상호작용_Json변환용.xlsx')

# 2. 6개의 열 이름 지정 (엑셀의 실제 열 순서와 맞춰주세요)
# 만약 엑셀의 열 순서가 다르다면 아래 리스트의 순서를 바꾸시면 됩니다.
df.columns = ['name', 'category', 'schedule', 'dosage', 'caution', 'description']

# 3. 객체 배열로 변환
data_list = df.to_dict(orient='records')

# 4. JSON 파일 저장
with open('userMedicationData.json', 'w', encoding='utf-8') as f:
    json_string = json.dumps(data_list, ensure_ascii=False, indent=4)
    # JS 변수 선언문 형식으로 저장
    f.write(f"const userMedicationData = {json_string};")

print("변환이 완료되었습니다!")
```
